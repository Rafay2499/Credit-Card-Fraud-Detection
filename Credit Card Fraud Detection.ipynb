```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score, confusion_matrix

def evaluate_model(y_true, y_predict):
    
    score = {
        "Precison" : precision_score(y_true,y_predict),
        "recall" : recall_score(y_true,y_predict),
        "accuracy" : accuracy_score(y_true,y_predict),
        "f1 score" : f1_score(y_true,y_predict)
        
    }
    print("--------Confusion Matrix---------")
    print(confusion_matrix(y_true,y_predict))
    print("--------------------------------")
    
    
    return score
```

```python
evaluate_model(Y_test,y_predict)
```

```python
import seaborn as sns

conf_matrix= confusion_matrix(Y_test,y_predict)

def plot_confmatrix(conf_matrix):
    fig, ax = plt.subplots(figsize=(3, 3))
    ax = sns.heatmap(conf_matrix, annot=True, cbar=False)
   
    plt.show()

# Assuming you have already defined `conf_matrix` and `Y_test` and `y_predicted` before this code
plot_confmatrix(conf_matrix)
## blacked area is confused matrix
```

# Tuning Hyperparameters

```python
from sklearn.model_selection import RandomizedSearchCV

rs_params ={
    "n_estimators" : np.arange(10,100,10),
    "max_depth" : [None, 3,5,10],
    "min_samples_split" : np.arange(2,20,2),
    "min_samples_leaf" : np.arange(1,20,2),
    "max_features" :[0.5,1.0,'Sqrt'],
    "max_samples" : [1000]
}

model_rs_LR = RandomizedSearchCV(RandomForestClassifier(),
                             param_distributions =rs_params,
                             n_iter = 20,
                             cv =5,
                             verbose =True)

model_rs_LR.fit(X_train,Y_train)
```

```python
model_rs_LR.best_params_
```

```python
# training the best parameters
best_params = model_rs_LR.best_params_

best_model = RandomForestClassifier(n_jobs =-1, random_state = 42, ** best_params)
best_model.fit(X_train,Y_train)

```
